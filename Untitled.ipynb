```python
from minsearch import Index
```

```python
index = Index(
    text_fields = ['topic', 'content'],
    keyword_fields=['tags']
)
```

```python
index.fit(data)
```

```python
index.search('Do you know anything about Transformers')
```

```python
import os
from groq import Client
```

```python
client = Client(api_key=os.environ.get("API_KEY"))
```

```python
def build_prompt(query, search_result):
    prompt_template = """
You are a course assistant bot. The only source of truth is the CONTEXT provided from MinSearch.

- Return the answer **exactly as it appears in the CONTEXT**, without rephrasing, summarizing, or adding anything.
- Do not generate or invent information that is not in the CONTEXT (no hallucinations).
- If the CONTEXT does not contain an answer, respond exactly with: "I don't have info on that".
- Do not add explanations, formatting, or extra text outside of the extracted answer.

CONTEXT:
{context}
"""

    context = ""
    for topics in search_result:
        context = context + f"topic: {topics['topic']}\n content: {topics['content']}\n\n"
        prompt = prompt_template.format(question = query, context = context).strip()
        return prompt
```

```python
query = "What is overfitting?"
search_result = index.search(query)
build_prompt(query, search_result)
```

```python
def llm(prompt):
    response = client.chat.completions.create( 
        model='llama-3.1-8b-instant', 
        messages= [{ 'role': 'user', 'content':prompt }]
    )
    return response.choices[0].message.content
```

```python
def search(query):
    return index.search(query)
```

```python
def ask(query):
    search_re = search(query)
    promptt = build_prompt(query, search_re)
    return llm(promptt)
```

```python
print(ask('What is gradient boosting'))
```
